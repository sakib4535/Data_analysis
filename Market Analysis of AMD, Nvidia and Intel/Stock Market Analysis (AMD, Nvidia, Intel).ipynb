```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amd = pd.read_csv("AMD.csv", header=0, parse_dates=True, index_col="date")
nvda = pd.read_csv("Nvidia.csv", header=0, parse_dates=True, index_col="date")
intel = pd.read_csv("Intel.csv", header=0, parse_dates=True, index_col="date")
```

```python
amd["open"].plot(figsize=(13,8))
nvda["open"].plot(figsize=(13,8))
intel["open"].plot(figsize=(13,8))
plt.xlabel("Timeline")
plt.ylabel("Opening Price(in Dollar)")
```

```python
#Volumn Trade Interpretations 
```

```python
amd["volume"].plot(figsize=(13,8), label="AMD")
nvda["volume"].plot(figsize=(13,8), label="Nvidia")
intel["volume"].plot(figsize=(13,8), label="Intel")
plt.xlabel("Timeline")
plt.ylabel("Opening Price(in Dollar)")
plt.legend();
```

```python
def main():
    
    stock = pd.read_csv("Amd.csv")
    fig = plt.figure(figsize=(13,8))
    
    ax = fig.add_subplot(1, 1, 1)
    stock["date"] = pd.to_datetime(stock["date"])
    ax.plot(stock["date"], stock["open"], linestyle="solid")
    plt.xlabel("Timeline")
    plt.ylabel("Price(in Dollar)")
    plt.title("Opening Price Data(AMD)")
    return

main()
```

```python
def main():
    
    stock = pd.read_csv("Nvidia.csv")
    fig = plt.figure(figsize=(13,8))
    
    ax = fig.add_subplot(1, 1, 1)
    stock["date"] = pd.to_datetime(stock["date"])
    ax.plot(stock["date"], stock["open"], linestyle="solid", color="red")
    plt.xlabel("Timeline")
    plt.ylabel("Price(in Dollar)")
    plt.title("Opening Price Data(Nvidia)")
    return

main()
```

```python
def main():
    
    stock = pd.read_csv("Intel.csv")
    fig = plt.figure(figsize=(13,8))
    
    ax = fig.add_subplot(1, 1, 1)
    stock["date"] = pd.to_datetime(stock["date"])
    ax.plot(stock["date"], stock["open"], linestyle="solid", color="green")
    plt.xlabel("Timeline")
    plt.ylabel("Price(in Dollar)")
    plt.title("Opening Price Data(Intel)")
    return

main()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline

f, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
#f.gca().xaxis.set_major_formatter(mdates.YearLocator())

ax[0,0].plot(amd.index, amd["close"], color="r")
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title("AMD")

ax[0,1].plot(nvda.index, nvda["close"], color="g")
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title("Nvidia")

ax[1,0].plot(intel.index, intel["close"], color="b")
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title("Intel")
```
